```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cost_of_living = "./Resources/cost_of_living.csv"
```

```python
col_df = pd.read_csv(cost_of_living)
```

```python
col_df.head()
```

```python
avg_1b = round(col_df['Total Cost for 1 BHK '].mean())
avg_3b = round(col_df['Total Cost of 3 BHK '].mean())
avg_u = round(col_df['Utilities Per Month'].mean())
avg_avg = round(col_df['Average Cost of Living '].mean())
```

```python
#scatter 1 bedroom
ax = col_df[['City Name','Total Cost for 1 BHK ']].plot(kind='bar', title ="1 bedroom", figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("city", fontsize=12)
ax.set_ylabel("cost", fontsize=12)
ax.set_title("1 bedroom cost", fontsize=12)

x_axis = np.arange(len(col_df))
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, col_df["City Name"], rotation="vertical")

plt.axhline(y=avg_1b, color='r', linestyle='-')

plt.show()
plt.savefig("./Resources/cst1bdroom.png")
```

```python
plt.tight_layout()
plt.savefig("./Resources/cost1bdroom.png")
```

```python
#scatter 3 bedroom
ax = col_df[['City Name','Total Cost of 3 BHK ']].plot(kind='bar', title ="1 bedroom", figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("city", fontsize=12)
ax.set_ylabel("cost", fontsize=12)
ax.set_title("3 bedroom cost", fontsize=12)

x_axis = np.arange(len(col_df))
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, col_df["City Name"], rotation="vertical")

plt.axhline(y=avg_3b, color='r', linestyle='-')

plt.show()
```

```python
plt.tight_layout()
plt.savefig("./Resources/cost3bdroom.png")
```

```python
#scatter utility
ax = col_df[['City Name','Utilities Per Month']].plot(kind='bar', title ="1 bedroom", figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("city", fontsize=12)
ax.set_ylabel("cost", fontsize=12)
ax.set_title("Utilities", fontsize=12)

x_axis = np.arange(len(col_df))
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, col_df["City Name"], rotation="vertical")

plt.axhline(y=avg_u, color='r', linestyle='-')

plt.show()
```

```python
plt.tight_layout()
plt.savefig("./Resources/utility.png")
```

```python
#scatter avg
ax = col_df[['City Name','Average Cost of Living ']].plot(kind='bar', title ="Average Cost of living", figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("city", fontsize=12)
ax.set_ylabel("cost", fontsize=12)
ax.set_title("Average Cost of living", fontsize=12)

x_axis = np.arange(len(col_df))
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, col_df["City Name"], rotation="vertical")

plt.axhline(y=avg_avg, color='r', linestyle='-')

plt.show()
```

```python
plt.tight_layout()
plt.savefig("./Resources/avgcol.png")
```
